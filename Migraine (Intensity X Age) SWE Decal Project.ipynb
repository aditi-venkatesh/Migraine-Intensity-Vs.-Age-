```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
migraine = pd.read_csv('data.csv')
migraine
```

```python
ages = list(migraine['Age'])
intensities = list(migraine['Intensity'])
intensities
```

```python
i_0 = []
i_1 = []
i_2 = []
i_3 = []

for i in range(0, len(ages)):
    if intensities[i] == 0:
        i_0.append(ages[i])
    elif intensities[i] == 1:
        i_1.append(ages[i])
    elif intensities[i] == 2:
        i_2.append(ages[i])
    else:
        i_3.append(ages[i])
```

```python
avg_ages = []
avg_age0 = np.mean(i_0)
avg_ages.append(avg_age0)
avg_age1 = np.mean(i_1)
avg_ages.append(avg_age1)
avg_age2 = np.mean(i_2)
avg_ages.append(avg_age2)
avg_age3 = np.mean(i_3)
avg_ages.append(avg_age3)
avg_ages
```

```python
mu = np.mean(intensities)
sigma = np.std(intensities)
print("Mean:", mu)
print("Standard deviation:", sigma)
```
